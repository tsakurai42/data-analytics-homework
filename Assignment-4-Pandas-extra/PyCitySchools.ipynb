```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
num_unique_schools = school_data_complete['school_name'].nunique()
num_unique_students = school_data_complete['student_name'].count()
total_budget = school_data['budget'].sum()
avg_math_score = school_data_complete['math_score'].mean()
avg_reading_score = school_data_complete['reading_score'].mean()

school_data_complete['overall_pass_rate'] = (school_data_complete['reading_score'] + school_data_complete['math_score']) / 2
#overall_passing_rate = (avg_math_score+avg_reading_score)/2

passing_math_students = school_data_complete.loc[school_data_complete['math_score'] >= 70]['Student ID'].count()
pass_math_pct = passing_math_students / num_unique_students

passing_reading_students = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['Student ID'].count()
pass_read_pct = passing_reading_students / num_unique_students

passing_overall_students = school_data_complete.loc[school_data_complete['overall_pass_rate'] >= 70]['Student ID'].count()
pass_overall_pct = passing_overall_students / num_unique_students

district_summary_df = pd.DataFrame({"Total Schools":[num_unique_schools], "Total Students":["{:,.0f}".format(num_unique_students)],
                                    "Total Budget":["${:,.2f}".format(total_budget)],"Average Math Score":avg_math_score,
                                   "Average Reading Score":[avg_reading_score], "% Passing Math":["{:.3%}".format(pass_math_pct)],
                                   "% Passing Reading":["{:.3%}".format(pass_read_pct)],"% Overall Passing Rate":["{:.3%}".format(pass_overall_pct)]})
district_summary_df

#overall passing is incorrect because i calculated read+math  /2 for each student then did loc > 70 instead of just averaging
#school_data_complete.head(25)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data_complete.head()
school_grouped = school_data_complete.groupby(by='school_name')
school_num_students = school_grouped['Student ID'].count()


type_list = []
for _ in school_num_students.index:
    type_list.append(school_data_complete['type'][school_data_complete['school_name'] == _].iloc[0])

school_performance_df = pd.DataFrame(school_num_students)
school_performance_df.rename(columns={'Student ID':'Total Students'},inplace=True)
school_performance_df['Type'] = type_list
school_performance_df = school_performance_df[['Type','Total Students']]

budget_list = []
for _ in school_num_students.index:
    budget_list.append(school_data_complete['budget'][school_data_complete['school_name'] == _].iloc[0])

school_performance_df['Budget'] = budget_list
school_performance_df["Per Student Budget"] = school_performance_df['Budget']/school_performance_df['Total Students']

school_mean_math = school_grouped['math_score'].mean()
school_performance_df['Avg Math Score'] = school_mean_math

school_mean_read = school_grouped['reading_score'].mean()
school_performance_df['Avg Reading Score'] = school_mean_read

pass_math_by_school = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count()
school_performance_df['%Passing Math'] = pass_math_by_school/school_num_students

pass_read_by_school = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count()
school_performance_df['%Passing Reading'] = pass_read_by_school/school_num_students
school_performance_df['%Overall Passing'] = (school_performance_df["%Passing Math"] + school_performance_df["%Passing Reading"]) / 2
school_performance_df['%Passing Math'] = school_performance_df['%Passing Math'].map("{:.2%}".format)
school_performance_df['%Passing Reading'] = school_performance_df['%Passing Reading'].map("{:.2%}".format)

school_performance_df['Budget'] = school_performance_df['Budget'].map('{:,.0f}'.format)

top_school_performance_df = school_performance_df.sort_values(by='%Overall Passing', ascending=False)
bottom_school_performance_df = school_performance_df.sort_values(by='%Overall Passing', ascending=True)

top_school_performance_df['%Overall Passing'] = top_school_performance_df['%Overall Passing'].map("{:.2%}".format)
bottom_school_performance_df['%Overall Passing'] = bottom_school_performance_df['%Overall Passing'].map("{:.2%}".format)

top_school_performance_df["Per Student Budget"] = top_school_performance_df["Per Student Budget"].map("${:.2f}".format)
bottom_school_performance_df["Per Student Budget"] = bottom_school_performance_df["Per Student Budget"].map("${:.2f}".format)

top_school_performance_df.head()
```
