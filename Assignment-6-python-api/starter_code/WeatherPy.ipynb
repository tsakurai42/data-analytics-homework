### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
owm_url = f'http://api.openweathermap.org/data/2.5/weather?APPID={api_key}&q='

#print(json.dumps(requests.get(owm_url+cities[7]).json(),indent=4))
city_jsons = []
print("Beginning Data Retrieval\n------------------------")
# for i in range(len(cities)):
#     city_json = requests.get(owm_url+cities[i]).json()
#     if city_json['cod'] == 200:
#         print(f"Processing Record {i} of {num_cities} | {city_json['name']} ")
#         city_jsons.append(city_json)
#     else:
#         print(f"{cities[i]} not found, skipping")


#combining the create city list and fetch data steps to make it easier to make sure to get 500 unique cities that are available
#in OWM.

i = 1             
while len(city_jsons) < num_cities:
    lat = np.random.uniform(low=-90.000, high=90.000)
    lng = np.random.uniform(low=-180.000, high=180.000)
    city = citipy.nearest_city(lat, lng).city_name
    if city not in cities:
        cities.append(city)
        city_json = requests.get(owm_url+city).json()
        if city_json['cod'] == 200: #success response
            print(f"Processing Record {i} of {num_cities} | {city_json['name']} ")
            city_jsons.append(city_json)
            i += 1
        else:
            print(f"{city} not found, retrying")
    else:
        print(f"Repeat city, retrying")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create a dataframe
cityweather_df = pd.DataFrame(columns=["City","Cloudiness","Country","Date","Humidity","Latitude","Longitude","MaxTemp","WindSpeed"])
for citydata in city_jsons:
    cityweather_df.loc[len(cityweather_df)] = [citydata['name'], citydata['clouds']['all'], citydata['sys']['country'], citydata['dt'], citydata['main']['humidity'], citydata['coord']['lat'], citydata['coord']['lon'], citydata['main']['temp_max']-273.15, citydata['wind']['speed']]

cityweather_df.set_index('City',inplace=True)
cityweather_df.to_csv("City Weather Data.csv",index=False)
cityweather_df.count()
```

#### Latitude vs. Temperature Plot

```python
fig,ax1 = plt.subplots()
ax1.scatter(cityweather_df['Latitude'],cityweather_df['MaxTemp'],alpha=0.5)
ax1.grid(axis="both")
ax1.set_xlim(-90,90)
ax1.set_title(f"Max Temp vs Latitude ({date_read})")
ax1.set_xlabel("Latitude")
ax1.set_ylabel("Max Temp (Celcius)")

ax2 = ax1.twinx()
ax2.set_ylabel("Max Temp (Fahrenheit)")
mini,maxi = ax1.get_ylim()
ax2.set_ylim(9/5*mini+32,9/5*maxi+32)
plt.savefig("Temp vs Latitude.png")
plt.show
```

```python
#To be honest, for THIS case, I think having the independent variable (latitude) on the y axis makes more sense
#if only because everyone associates latitude with the map and it makes more intuitive sense to think going up and down the
#y-axis equates to going north or south in the world.

fig,ax1 = plt.subplots()
ax1.scatter(cityweather_df['MaxTemp'],cityweather_df['Latitude'],alpha=0.5)
ax1.grid(axis="both")
ax1.set_ylim(-90,90)
ax1.set_title(f"Max Temp vs Latitude ({date_read})")
ax1.set_ylabel("Latitude")
ax1.set_xlabel("Max Temp (Celcius)")

ax2 = ax1.twiny()
ax2.set_xlabel("Max Temp (Fahrenheit)")
mini,maxi = ax1.get_xlim()
ax2.set_xlim(9/5*mini+32,9/5*maxi+32)
plt.savefig("Temp vs Latitude (rot).png",bbox_inches="tight")
plt.show
```

```python
# As expected, Temperature is highest near the equator. It also falls off as you increase in Latitude either direction
#It looks like there is far more data in the northern hemisphere than southern
```

```python
#testing openweathermap's bbox function to grab a bunch of cities within a box(box of -180,-90 to 180,90, AKA the whole world)
#it does not state how many cities it can return.
mass_weather_dump = requests.get(f"http://api.openweathermap.org/data/2.5/box/city?bbox=-180,-90,180,90&APPID={api_key}").json()
```

```python
mass_weather_dump_df = pd.DataFrame(columns=["Name","Lat","Long","Temp","Time"])
for eachcity in mass_weather_dump['list']:
    mass_weather_dump_df.loc[len(mass_weather_dump_df)] = [eachcity['name'],eachcity['coord']['Lat'],eachcity['coord']['Lon'],eachcity['main']['temp_max'],eachcity['dt']]

mass_dump_date_read = datetime.fromtimestamp(mass_weather_dump_df['Time'].max()).strftime('%Y-%m-%d')

plt.figure(figsize=(10,6))
plt.scatter(mass_weather_dump_df['Long'],mass_weather_dump_df['Lat'],c=mass_weather_dump_df["Temp"],cmap='rainbow')
cbar = plt.colorbar()
cbar.set_label("Temp (C)")
plt.title(f"Temp vs Latitude and Longitude ({mass_dump_date_read})")
plt.ylim(-90,90)
plt.xlim(-180,180)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.hlines(0,-180,180)
plt.savefig("Temperature Map (method2).png")
plt.show
```
